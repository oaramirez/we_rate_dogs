#### Now let's read that txt file and make into a DF

```python
df_tweets = pd.DataFrame()
df_tweets.info()
```

Useful info:
    
- https://www.geeksforgeeks.org/read-a-file-line-by-line-in-python/
- https://stackabuse.com/reading-and-writing-json-to-a-file-in-python/

```python
filename = 'tweet_json.txt'
with open(filename, mode='r') as file:
    lines = file.readlines()
    for line in lines:
        data = json.loads(line)
        tweet_id = data['id']
        favorite_count = data['favorite_count']
        retweet_count = data['retweet_count']
        try:
            img_uri = data['entities']['media'][0]['media_url_https']
        except:
            img_uri = ''
#             print(f'no image on tweet {id_value}')
        finally:
#             print(f'completed tweet {id_value}')
            values = {
#                 'tweet_id' : str(tweet_id),
                'tweet_id' : str(tweet_id),
                'favorite_count' : int(favorite_count),
                'retweet_count' : int(retweet_count),
                'img_uri' : img_uri
            }
            df_tweets = df_tweets.append(values, ignore_index = True)
```
